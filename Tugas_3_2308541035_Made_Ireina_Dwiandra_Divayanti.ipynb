```python
#2.10

suhu_35 = float(input("Berapa suhu ruangan tersebut?"))
celcius_kelvin_35 = input("suhu dalam celcius atau kelvin?")

if celcius_kelvin_35.lower() == "celcius":
    suhu_35 = suhu_35+273
    print("Jadi, suhu dalam kelvin adalah " + str(suhu_35) + " Kelvin")
elif celcius_kelvin_35.lower() == "kelvin":
    suhu_35 = suhu_35-273
    print("Jadi, suhu dalam celcius adalah " + str(suhu_35) + " Celcius")
    
```
