```python
import matplotlib.pyplot as plt
import numpy as np


a_35 = 1    #x1
b_35 = 1    #y1
c_35 = 2   #x2
d_35 = 2    #y2
m1_35 = (d_35-b_35)/(c_35-a_35)
print("Gradien dari garis pertama adalah " + str(m1_35))
pgl1_35 = print("Maka persamaan garis satu adalah " + str(m1_35) + "x +" + str(-m1_35*a_35+b_35))


e_35 = 2
f_35 = -5
g_35 = 1
h_35 = -3
m2_35 = (h_35-f_35)/(g_35-e_35)
print("Gradien dari garis dua adalah " + str(m2_35))
pgl2_35 = print("Maka persamaan garis dua adalah " + str(m2_35) + "x +" + str(-m2_35*e_35+f_35))

if m1_35 == m2_35:
    print("Garis memiliki banyak solusi atau tidak memiliki solusi")
else:
    tpx_35 = (m1_35*a_35-b_35-m2_35*e_35+f_35)/(m1_35-m2_35)
    tpy_35 = m2_35*(tpx_35-e_35)+f_35
    print("Kedua garis tersebut berpotongan di ({},{})".format(tpx_35,tpy_35))



x_35=np.arange(-3,5,0.2)
p1_35=plt.plot(x_35,m1_35*(x_35-a_35)+b_35)
p2_35=plt.plot(x_35,m2_35*(x_35-e_35)+f_35)
plt.plot(tpx_35,tpy_35,marker='o',markersize=6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_35 = np.arange(0, 3*np.pi, 0.1)
y_35 = np.sin(x_35)

plt.plot(x_35, y_35)
plt.show()
```

```python
#grafik untuk y = x^2

import numpy as np
import matplotlib.pyplot as plt

x_35 = np.arange(-2, 3, 0.01)
y_35 = x_35 * x_35

plt.plot(x_35, y_35)
plt.show()
```

```python
#grafik fungsi polinomial

import numpy as np
import matplotlib.pyplot as plt

x_35 = np.arange(-2, 3, 0.01)
y_35 = np.power(x_35, 3) + np.power(x_35, 2) + 3*x_35 +1

plt.plot(x_35, y_35)
plt.show()
```
