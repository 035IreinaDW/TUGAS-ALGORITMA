```python
#2.18

kode_morse_35= {
    "A": ".-",
    "B": "-...",
    "C": "-.-.",
    "D": "-..",
    "E": ".",
    "F": "..-.",
    "G": "--.",
    "H": "....",
    "I": "..",
    "J": ".---",
    "K": "-.-",
    "L": ".-..",
    "M": "--",
    "N": "-.",
    "O": "---",
    "P": ".--.",
    "Q": "--.-",
    "R": ".-.",
    "T": "-",
    "U": "..-",
    "V": "..-",
    "W": ".--",
    "X": "-..-",
    "Y": "-.--",
    "Z": "--.."
}

pesan_35 = input('Masukkan pesan: ')
rahasia_35 = pesan_35.upper()
pesan_morse_35 = ""
for isi in rahasia_35:
    pesan_morse_35 += kode_morse_35.get(isi) + ""
print(pesan_morse_35)
```
