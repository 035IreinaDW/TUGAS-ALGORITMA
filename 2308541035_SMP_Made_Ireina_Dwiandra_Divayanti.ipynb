```python
import cmath

#Diketahui x yang memenuhi persamaan x**2 + x - 1 = 0
#Tentukan nilai n yang memenuhi x**2024 - x**2025 = (x(1-x))**n

#Dicari nilai x terlebih dahulu
x_35 = (-1+cmath.sqrt(5))/2
print("Didapat nilai x adalah {}".format(x_35))

#x^2024 - x^2025 = x^2024(1-x)
#Diketahui log ab = log a + log b
pembilang_35= 2024*cmath.log(x_35) + cmath.log(1-x_35)

#(x(1-x))
persamaan_35 = x_35 * (1-x_35)
penyebut_35 = cmath.log(persamaan_35)
#Nilai n
n_35 = pembilang_35 / penyebut_35
print("Maka nilai n adalah {}".format(n_35.real))
```
