```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

a_35 = float(input("Masukkan nilai a = "))
b_35 = float(input("Masukkan nilai b = "))
c_35 = float(input("Masukkan nilai c = "))
pers_35 = print("Maka persamaan kuadratnya adalah f(x) = {}x^2 + {}x + {}".format(a_35, b_35, c_35))

if a_35>0:
    print("Grafik akan membentuk parabola terbuka ke atas")
elif a_35<0:
    print("Grafik akan membentuk parabola terbuka ke bawah")
else:
    print("Grafik tidak dapat ditunjukkan")
print()
    
    
print("Dicari determinan dari persamaan tersebut terlebih dahulu")

D_35 = (b_35**2)-(4*a_35*c_35)

if D_35 < 0:
    print("Didapat diskriminannya kurang dari 0 yakni " + str(D_35))
    print("Persamaan kuadrat tersebut memiliki dua akar khayal(tidak real/imajiner)")

if D_35 == 0:
    print("Didapat diskriminannya sama dengan nol")
    print("Persamaan kuadrat tersebut memiliki duar akar nyata(real) yang sama/kembar")
    
if D_35 > 0:
    print("Didapat diskriminannya lebih dari nol yakni " + str(D_35))
    print("Persamaan kuadrat tersebut memiliki dua akar nyata(real) yang berbeda")
    
else:
    print("Dilanjutkan mencari akar-akar persamaan kuadrat tersebut")
print()

if a_35!=0:
    akar1_35 = ((-b_35+cmath.sqrt(D_35))/(2*a_35))
    akar2_35 = ((-b_35-cmath.sqrt(D_35))/(2*a_35))
    print("Jadi, akar-akar dari persamaan tersebut adalah {} dan {}".format(akar1_35,akar2_35))
if a_35==0:
    print("Tidak termasuk persamaan kuadrat karena nilai a sama dengan 0! ")
else:
    print("Dilanjutkan mencari titik puncak")
print()
    
xp_35 = -b_35/(2*a_35)
yp_35 = (a_35*(xp_35**2))+(b_35*xp_35)+c_35
print("Titik puncak dari persamaan tersebut adalah ({},{})".format(xp_35,yp_35))

tpx1_35 = akar1_35
tpx2_35 = akar2_35
print("Titik potong di sumbu x adalah ({},0) dan ({},0)".format(tpx1_35,tpx2_35))
tpy_35 = c_35
print("Titik potong sumbu y = ({},0)".format(tpy_35))

simetri_35 = -b_35/(2*a_35)
print("Sumbu simetri dari persamaan tersebut adalah " + str(simetri_35))

x_35 = np.linspace(xp_35 - 5, xp_35 + 5)
y_35 = a_35*(x_35**2) + b_35*x_35 + c_35
plt.plot(x_35,y_35, color="pink",label="y = a_35*(x**2) + b_35*x + c_35")
plt.legend()
plt.scatter(x_35,y_35, color="pink",label="y = a_35*(x**2) + b_35*x + c_35")
plt.title("Grafik Persamaan Kuadrat")
plt.xlabel("Sumbu x")
plt.ylabel("Sumbu y")
plt.show() 
```
